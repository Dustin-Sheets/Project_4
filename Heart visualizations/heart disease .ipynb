```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("heart.csv")
df.head(1026)
```

```python
data = {
    'Age': [35,40,45,50,55,60,65,70],
    'Sex': ['Male', 'Female', 'Male', 'Female','Male', 'Female', 'Male', 'Female'],
    'ChestPain': ['Mild','Moderate', 'Severe','Mild','Moderate', 'Severe','Mild','Moderate']
}

df = pd.DataFrame(data)

grouped = df.groupby(['Sex', 'ChestPain']).size().unstack()

# Plotting the grouped bar chart
grouped.plot(kind='bar', color=['yellow','orange','red'], stacked=False)  
plt.ylabel('Pain Level')
plt.title('Chart of Sex and Chest Pain')
plt.legend(title='Chest Pain')
plt.show()
```
